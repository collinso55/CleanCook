```python
#ooop
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
# Set up plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
# Load the dataset
print("Loading dataset...")
df = pd.read_csv('../data/processed/fuel_consumption_master_dataset_2010_2024.csv')

print(f"Dataset shape: {df.shape}")
print(f"Date range: {df['date'].min()} to {df['date'].max()}")

# Display basic info
print("\nDataset Info:")
print(df.info())

# Display first few rows
print("\nFirst 5 rows:")
display(df.head())

# Display basic statistics
print("\nBasic Statistics:")
display(df.describe())
```

```python
# =============================================================================
# 1. DATA QUALITY ASSESSMENT
# =============================================================================

print("="*80)
print("1. DATA QUALITY ASSESSMENT")
print("="*80)

# Check for missing values
print("\nMissing Values Analysis:")
missing_data = df.isnull().sum()
missing_percent = (missing_data / len(df)) * 100
missing_info = pd.DataFrame({
    'Missing Count': missing_data,
    'Missing Percentage': missing_percent
})
display(missing_info[missing_info['Missing Count'] > 0])

# Check for duplicates
print(f"\nDuplicate rows: {df.duplicated().sum()}")

# Check data types
print("\nData Types:")
print(df.dtypes)

# Convert date to datetime format
print("\nConverting date to datetime format...")
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date').reset_index(drop=True)
```

```python

# =============================================================================
# 2. TEMPORAL ANALYSIS
# =============================================================================

print("\n" + "="*80)
print("2. TEMPORAL ANALYSIS")
print("="*80)

# Check if dates are continuous
date_range = pd.date_range(start=df['date'].min(), end=df['date'].max(), freq='D')
print(f"Expected days: {len(date_range)}")
print(f"Actual days in dataset: {len(df)}")
print(f"Missing days: {len(date_range) - len(df)}")

# Identify missing dates
all_dates = pd.DataFrame({'date': date_range})
missing_dates = all_dates[~all_dates['date'].isin(df['date'])]
if not missing_dates.empty:
    print(f"\nMissing dates found: {len(missing_dates)}")
    print("Sample missing dates:")
    display(missing_dates.head())
else:
    print("\nNo missing dates found - dataset is continuous!")
```

```python

# =============================================================================
# 3. TARGET VARIABLE ANALYSIS
# =============================================================================

print("\n" + "="*80)
print("3. TARGET VARIABLE ANALYSIS (woodfuel_production_m3)")
print("="*80)

# Basic statistics of target variable
target_stats = df['woodfuel_production_m3'].describe()
print("Target Variable Statistics:")
display(target_stats)

# Check for outliers in target variable using IQR method
Q1 = df['woodfuel_production_m3'].quantile(0.25)
Q3 = df['woodfuel_production_m3'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['woodfuel_production_m3'] < lower_bound) | 
             (df['woodfuel_production_m3'] > upper_bound)]
print(f"Number of outliers in target variable (IQR method): {len(outliers)}")
print(f"Outlier percentage: {(len(outliers)/len(df))*100:.2f}%")

# Plot target variable distribution
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(df['date'], df['woodfuel_production_m3'])
plt.title('Woodfuel Production Over Time')
plt.xlabel('Date')
plt.ylabel('Production (m³)')
plt.xticks(rotation=45)

plt.subplot(2, 2, 2)
plt.hist(df['woodfuel_production_m3'], bins=50, edgecolor='black')
plt.title('Distribution of Woodfuel Production')
plt.xlabel('Production (m³)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
df['woodfuel_production_m3'].plot(kind='box')
plt.title('Boxplot of Woodfuel Production')

plt.subplot(2, 2, 4)
# Monthly aggregation
monthly_avg = df.groupby('month')['woodfuel_production_m3'].mean()
plt.bar(monthly_avg.index, monthly_avg.values)
plt.title('Average Monthly Production')
plt.xlabel('Month')
plt.ylabel('Average Production (m³)')

plt.tight_layout()
plt.show()
```

```python
# =============================================================================
# 4. SEASONALITY AND TREND ANALYSIS
# =============================================================================

print("\n" + "="*80)
print("4. SEASONALITY AND TREND ANALYSIS")
print("="*80)

# Analyze seasonal patterns
seasonal_analysis = df.groupby('season').agg({
    'woodfuel_production_m3': ['mean', 'std', 'count']
}).round(2)
print("Production by Season:")
display(seasonal_analysis)

# Yearly trend analysis
yearly_trend = df.groupby('year').agg({
    'woodfuel_production_m3': ['mean', 'sum', 'std']
}).round(2)
print("Yearly Trend Analysis:")
display(yearly_trend)

# Monthly patterns across years
monthly_patterns = df.pivot_table(
    values='woodfuel_production_m3', 
    index='month', 
    columns='year', 
    aggfunc='mean'
)
print("Monthly Patterns Across Years:")
display(monthly_patterns)

# Plot seasonal patterns
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
yearly_trend[('woodfuel_production_m3', 'mean')].plot(kind='bar')
plt.title('Average Yearly Production')
plt.xlabel('Year')
plt.ylabel('Average Production (m³)')

plt.subplot(2, 2, 2)
seasonal_analysis[('woodfuel_production_m3', 'mean')].plot(kind='bar')
plt.title('Average Production by Season')
plt.xlabel('Season')
plt.ylabel('Average Production (m³)')

plt.subplot(2, 2, 3)
# Weekly seasonality
df['day_of_week'] = df['date'].dt.dayofweek
weekly_pattern = df.groupby('day_of_week')['woodfuel_production_m3'].mean()
plt.bar(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'], weekly_pattern)
plt.title('Average Production by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Average Production (m³)')

plt.subplot(2, 2, 4)
# Quarterly pattern
quarterly_pattern = df.groupby('quarter')['woodfuel_production_m3'].mean()
plt.bar(['Q1', 'Q2', 'Q3', 'Q4'], quarterly_pattern)
plt.title('Average Production by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Average Production (m³)')

plt.tight_layout()
plt.show()
```

```python
# =============================================================================
# 5. FEATURE CORRELATION ANALYSIS
# =============================================================================

print("\n" + "="*80)
print("5. FEATURE CORRELATION ANALYSIS")
print("="*80)

# Select numerical features for correlation analysis
numerical_features = df.select_dtypes(include=[np.number]).columns.tolist()

# Remove identifier columns and highly correlated lag features
exclude_features = ['year', 'day_of_year', 'woodfuel_lag1', 'woodfuel_lag7', 
                   'woodfuel_7d_avg', 'woodfuel_30d_avg', 'temp_7d_avg', 
                   'temp_30d_avg', 'rainfall_7d_avg', 'rainfall_30d_cumulative']

analysis_features = [f for f in numerical_features if f not in exclude_features]

# Correlation matrix
correlation_matrix = df[analysis_features].corr()

# Plot correlation heatmap
plt.figure(figsize=(16, 14))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', 
            center=0, fmt='.2f', square=True)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()

# Top correlations with target variable
target_correlations = correlation_matrix['woodfuel_production_m3'].sort_values(ascending=False)
print("Top correlations with woodfuel production:")
display(target_correlations.head(15))
```

```python
# =============================================================================
# 6. FEATURE ENGINEERING AND SELECTION
# =============================================================================

print("\n" + "="*80)
print("6. FEATURE ENGINEERING AND SELECTION")
print("="*80)

# Create a copy for feature engineering
df_processed = df.copy()

# 6.1 Remove redundant features
print("Removing redundant features...")
redundant_features = [
    'woodfuel_lag1', 'woodfuel_lag7',  # Already captured in rolling features
    'temp_7d_avg', 'temp_30d_avg',     # Redundant with temperature_avg
    'rainfall_7d_avg', 'rainfall_30d_cumulative',  # Redundant with rainfall_mm
    'day_of_year',  # Redundant with date and month
    'season',  # Can be derived from month
]

# Keep only one seasonal indicator (remove one of the redundant season indicators)
if 'is_dry_season' in df_processed.columns and 'is_rainy_season' in df_processed.columns:
    redundant_features.append('is_rainy_season')  # Keep only dry season indicator

df_processed = df_processed.drop(columns=redundant_features)

# 6.2 Create new meaningful features
print("Creating new meaningful features...")

# Cyclical encoding for month and day_of_week
df_processed['month_sin'] = np.sin(2 * np.pi * df_processed['month']/12)
df_processed['month_cos'] = np.cos(2 * np.pi * df_processed['month']/12)
df_processed['day_sin'] = np.sin(2 * np.pi * df_processed['day_of_week']/7)
df_processed['day_cos'] = np.cos(2 * np.pi * df_processed['day_of_week']/7)

# Weather interaction features
df_processed['temp_rain_interaction'] = df_processed['temperature_avg'] * df_processed['rainfall_mm']
df_processed['humidity_wind_interaction'] = df_processed['humidity_avg'] * df_processed['wind_speed_kmh']

# Seasonal aggregates (rolling windows)
df_processed['production_7d_rolling_mean'] = df_processed['woodfuel_production_m3'].rolling(window=7).mean()
df_processed['production_30d_rolling_mean'] = df_processed['woodfuel_production_m3'].rolling(window=30).mean()
df_processed['production_7d_rolling_std'] = df_processed['woodfuel_production_m3'].rolling(window=7).std()

# Lag features (different from the redundant ones)
df_processed['production_lag_30'] = df_processed['woodfuel_production_m3'].shift(30)
df_processed['production_lag_365'] = df_processed['woodfuel_production_m3'].shift(365)

# 6.3 Handle missing values from rolling calculations
print("Handling missing values from feature engineering...")
df_processed = df_processed.fillna(method='bfill')  # Backfill for initial missing values

# 6.4 Feature selection based on correlation and domain knowledge
print("Performing final feature selection...")

# Keep features with reasonable correlation and domain relevance
selected_features = [
    # Temporal features
    'month_sin', 'month_cos', 'day_sin', 'day_cos', 'is_weekend', 'is_dry_season',
    
    # Weather features
    'temperature_avg', 'rainfall_mm', 'humidity_avg', 'wind_speed_kmh',
    'temp_rain_interaction', 'humidity_wind_interaction',
    
    # Production trends
    'production_7d_rolling_mean', 'production_30d_rolling_mean', 
    'production_7d_rolling_std', 'production_lag_30', 'production_lag_365',
    
    # Target variable
    'woodfuel_production_m3'
]

# Ensure all selected features exist in the dataframe
final_features = [f for f in selected_features if f in df_processed.columns]
df_final = df_processed[final_features + ['date', 'year', 'month']]  # Keep identifiers for analysis

print(f"Final dataset shape: {df_final.shape}")
print(f"Selected features: {len(final_features)}")
```

```python
# =============================================================================
# 7. FINAL DATA QUALITY CHECK
# =============================================================================

print("\n" + "="*80)
print("7. FINAL DATA QUALITY CHECK")
print("="*80)

# Check for missing values in final dataset
final_missing = df_final.isnull().sum()
print("Missing values in final dataset:")
display(final_missing[final_missing > 0])

# Check data types
print("\nFinal dataset data types:")
print(df_final.dtypes)

# Verify no infinite values
print(f"\nInfinite values: {np.isinf(df_final.select_dtypes(include=[np.number])).sum().sum()}")
```

```python

# =============================================================================
# 8. SAVE PROCESSED DATASET
# =============================================================================

print("\n" + "="*80)
print("8. SAVING PROCESSED DATASET")
print("="*80)

# Save the processed dataset
output_filename = 'fuel_consumption_processed_dataset.csv'
df_final.to_csv(output_filename, index=False)
print(f"Processed dataset saved as: {output_filename}")
print(f"Final dataset shape: {df_final.shape}")
```
